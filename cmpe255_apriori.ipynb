#frequent pattern mining Apriori Algorithm

```python
# Importing libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
# data Initialization
apriori_data = [['Tommy', 'Ben', 'Joe', 'John', 'Alex', 'Charles'],
           ['John', 'Tommy', 'Charles', 'Joe', 'Emily', 'Anna'],
           ['Ben', 'John', 'Tommy', 'Joe'],
           ['John', 'Alex', 'Joe', 'Tommy', 'Charles'],
           ['Ben', 'Alex', 'Chris', 'Tommy', 'Anna', 'Emily']]
```

```python
#data transformation
data1 = TransactionEncoder()
transformed_apriori_data = data1.fit(apriori_data).transform(apriori_data)
apriori_result = pd.DataFrame(transformed_apriori_data, columns=data1.columns_)
apriori_result
```

```python
#importing apriori
from mlxtend.frequent_patterns import apriori
apriori(apriori_result, min_support=0.3)
```

```python
apriori(apriori_result, min_support=0.7, use_colnames=True)
```

```python
# Importing libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
# data
apriori_Gdata = pd.read_csv("/content/Groceries_dataset.csv")
apriori_Gdata.itemDescription = ["".join(i.split(",")) for i in apriori_Gdata.itemDescription]
apriori_Gorder = apriori_Gdata.Member_number.unique()
```

```python
# Transforming the data
apriori_trans = [apriori_Gdata[apriori_Gdata.Member_number == i]["itemDescription"].tolist() for i in apriori_Gorder]
apriori_encoder = TransactionEncoder().fit(apriori_trans)
apriori_trans = apriori_encoder.transform(apriori_trans)
apriori_trans = pd.DataFrame(apriori_trans,columns=apriori_encoder.columns_)
```

```python
# Applying apriori algorithm
freq_apriori = apriori(apriori_trans,min_support=0.0095,use_colnames=True)

# Applying association rules
apriori_association_data = association_rules(freq_apriori,metric="confidence")
for item in apriori_association_data.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
```
